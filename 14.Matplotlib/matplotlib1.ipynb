Mathplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([5,10])
y = np.array([2,8])
plt.plot(x, y)
plt.show()
```

### Plot

```python
x = np.array([20, 30, 40, 50])
y = np.array([3, 2, 5, 4])
plt.plot(x, y)
plt.show()
```

```python
x = np.array([20, 30, 40, 50])
y = np.array([3, 2, 5, 4])
plt.plot(x, y, 'o')
plt.show()
```

### Marker

```python
x = np.array([20, 30, 40, 50])
y = np.array([3, 2, 5, 4])
plt.plot(x, y, marker = '*')
plt.show()
```

### Line

```python
x = np.array([20, 30, 40, 50])
y = np.array([3, 2, 5, 4])
plt.plot(x, y, ls = 'dashed', color = 'red', lw = '5')
plt.show()
```

### Labels And Title

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([38,45,36,40,51,63,50,70])

plt.plot(x,y)

font_style1 = {'family':'arial', 'color':'navy', 'size':16}
font_style2 = {'family':'arial', 'color':'green', 'size':14}

plt.xlabel("Test Number", fontdict=font_style2)
plt.ylabel("Test Score", fontdict=font_style2)
plt.title("Test Proformance Graph", fontdict=font_style1)

plt.show()
```

### Grid

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([38,45,36,40,51,63,50,70])

plt.plot(x,y)

font_style1 = {'family':'arial', 'color':'navy', 'size':16}
font_style2 = {'family':'arial', 'color':'green', 'size':14}

plt.xlabel("Test Number", fontdict=font_style2)
plt.ylabel("Test Score", fontdict=font_style2)
plt.title("Test Proformance Graph", fontdict=font_style1)

#grid
#plt.grid()
plt.grid(axis='x', color = 'green', ls = '--', lw = 4)
plt.grid(axis='y', color = 'red', ls = 'dotted')


plt.show()
```

### Subplot

```python
x = np.array([1,2,3,4,5])
y = np.array([38,45,36,40,51])
plt.subplot(1,2,1)
plt.plot(x, y)

x = np.array([1,2,3,4,5])
y = np.array([30,25,36,45,58])
plt.subplot(1,2,2)
plt.plot(x, y)

plt.show()
```

### Scatter Plot

```python
x = np.array([21,33,42,18,65,20,25,31,29,40,51])
y = np.array([65,59,66,61,78,60,67,58,60,61,68])

plt.scatter(x,y)

x = np.array([21,33,42,18,65,20,25,31,29,40,51])
y = np.array([61,58,62,59,75,57,61,52,63,59,65])

plt.scatter(x,y)


plt.show()
```

### Bar Graph

```python
x = np.array(['Jan','Fab','Mar', 'Apr', 'May'])
y = np.array([681, 1028, 986, 832, 901])

plt.bar(x,y)
#plt.barh(x,y,color='red')

plt.show()
```

### Histogram

```python
x = np.random.normal(200,10,100)

plt.hist(x)
plt.show()
```

### Pie Chart

```python
sections = np.array([25,10,35,30])
label = ["Production", "Marketing", "Transportation", "Others"] 
e = [0,0,0.1,0]
plt.pie(sections, labels=label, explode=e, shadow=True)
plt.legend()
plt.show()
```
